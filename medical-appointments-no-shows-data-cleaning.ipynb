```python
df_original = pd.read_csv("/kaggle/input/noshowappointments/KaggleV2-May-2016.csv")
df_original.shape
```

```python
df_original.isna().sum()
```

```python
df = df_original.copy()
df.shape
```

```python
df.head()
```

```python
#Applying snake_case to columns

df.columns = (
    df.columns
    .str.replace(r"([a-z0-9])([A-Z])", r"\1_\2", regex=True)
    .str.replace(r"[^\w]+", "_", regex=True)
    .str.lower()
)


df.head()
```

```python
df['patient_id'].apply(type).value_counts()
```

```python
df['patient_id'][0]
```

```python
#checking if decimal point patient_id exist or not!!?

decimal_point_check = df['patient_id'] % 1 != 0

df.loc[decimal_point_check, 'patient_id']
```

```python
#converting all patient_id from numerical to categorical

df["patient_id"] = df["patient_id"].astype(str).astype("category")

df['patient_id'].dtype
```

```python
df['appointment_id'].dtype
```

```python
#converting all appointment_id to categorical

df['appointment_id'] = df['appointment_id'].astype('category')

```

```python
df['appointment_id'].dtype
```

```python
df['appointment_id'].duplicated().value_counts()
```

```python
cols = ['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']

df[cols] = df[cols].astype(bool)
```

```python
df['scholarship'].dtype
```

```python
df = df.rename(columns ={'handcap': 'handicap'})

df.head()
```

```python
df.head()
```

```python
df['gender'].value_counts()
```

```python
df['age'].unique()
```

```python
#impossible age validation

age_valid_mask = (df['age'] >= 0) & (df['age'] <= 120)
df['age_valid'] = age_valid_mask.astype(bool)
```

```python
df.head()
```

```python
df['age_valid'].value_counts()
```

```python
df['scheduled_day'].apply(type).value_counts()
```

```python
#fixing datatype of scheduled_day
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'])

df['scheduled_day'].dtype
```

```python
df['appointment_day'].apply(type).value_counts()
```

```python
df['appointment_day'] = pd.to_datetime(df['appointment_day'])

df['appointment_day'].dtype
```

```python
#days between schedule and appointment

df['days_between_schedule_and_appointment'] = (
    (df['appointment_day'] - df['scheduled_day']).dt.days
)

df.head()
```

```python
df['days_between_schedule_and_appointment'].value_counts()
```

```python
df['same_day_appointment'] = (df['days_between_schedule_and_appointment'] == -1)


df.head()
```

```python
df['appointment_date_valid'] = (df['days_between_schedule_and_appointment'] >= -1)

df.head()
```

```python
df.sample(n=100, random_state=42).to_csv(
    "Medical Appointments No-Shows Cleaned_sample.csv",
    index=False
)
```
